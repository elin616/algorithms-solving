```python
cnt = int(input())
deq = ['None' for _ in range(cnt)]
front = 0
back = 0

for _ in range(cnt):
    want = input().split()
    if want[0] == 'push_front':
        num = int(want[1])
        deq[front] = num
        front -= 1
    
    elif want[0] == 'push_back':
        back += 1
        num = int(want[1])
        if back == 1 and front != 0:
            back = front+1
        deq[back] = num
    
    elif want[0] == 'pop_front':
        if front==back:
            print(-1)
        else : 
            print(deq[front+1])
            deq[front+1]='None'
            front += 1
    
    elif want[0] == 'pop_back':
        if front==back:
            print(-1)
        else:
            print(deq[back])
            deq[back]='None'
            back -= 1
    
    elif want[0] == 'size':
        print(back-front)
    
    elif want[0] == 'empty':
        if front == back:
            print(1)
        else:
            print(0)
    
    elif want[0] == 'front':
        if front==back:
            print(-1)
        else : 
            print(deq[front+1])

    elif want[0] == 'back':
        if front==back:
            print(-1)
        else : 
            print(deq[back])
    #print(want, deq)
```
